```python
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
Name=[]
Year_span=[]
Genre=[]
Run_time=[]
Ratings=[]
Votes=[]
```

```python
name=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]')
for i in name:
    Name.append(i.text)
year=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
for i in year:
    Year_span.append(i.text)
genre=driver.find_elements(By.XPATH,'//span[@class="genre"]')
for i in genre:
    Genre.append(i.text)
time=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
for i in time:
    Run_time.append(i.text)
ratings=driver.find_elements(By.XPATH,'//div[@class="ipl-rating-star small"]')
for i in ratings:
    Ratings.append(i.text)


```

```python
c=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/div/div[1]/div/div/form/label/input[1]')
c.click()
```

```python
a=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/div/div[1]/div/div/form/label/div/ul[3]/li[3]/a/div[3]/span[2]')
a.click()
```

```python
title=[]
description=[]
language=[]
contributors=[]
```

```python
start=0
end=50
for page in range(start,end):
    titles=driver.find_elements(By.XPATH,'//a[@class="v-align-middle"]')
    for i in titles:
        title.append(i.text)
    desc=driver.find_elements(By.XPATH,'//p[@class="mb-1"]')
    for i in desc:
        description.append(i.text)
    cont=driver.find_elements(By.XPATH,'//a[@class="Link--muted"]')
    for i in cont:
        contributors.append(i.text)
```
