```python
start=0
end=10
for page in range(start,end):
    try:
        names=driver.find_elements(By.XPATH,'//span[@class="fl ellipsis"]')
        for i in names:
            name.append(i.text)
    except NoSuchElementException:
        name.append('-')
    try:
        designations=driver.find_elements(By.XPATH,'//span[@class="ellipsis clr"]')
        for i in designations:
            designation.append(i.text)
    except NoSuchElementException:
        designation.append('-')
    try:
        companys=driver.find_elements(By.XPATH,'//i[@class="cmpIcn"]')
        for i in companys:
            company.append(i.text)
    except NoSuchElementException:
        company.append('-')
    try:
        skills=driver.find_elements(By.XPATH,'//div[@class="hireSec highlightable"]')
        for i in skills:
            skill.append(i.text)
    except NoSuchElementException:
        skill.append('-')
    try:
        locations=driver.find_elements(By.XPATH,'//small[@class="ellipsis"]')
        for i in locations:
            location.append(i.text)
    except NoSuchElementException:
        location.append('-')
    next= driver.find_element(By.XPATH,'/html/body/div[3]/div/div[2]/div/div[3]/button')
    next.click()
    
    
```
